```python
# Import libraries
import keras
from keras.datasets import mnist
```

```python
# Import datasets
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
# print(x_train.shape)
# print(y_train.shape)
# print(x_test.shape)
# print(y_test.shape)
```

```python
# Reshape and normalize
X_train = x_train.reshape(-1, 28*28)/255.0
X_test = x_test.reshape(-1, 28*28)/255.0
# print(X_train.shape)
# print(X_test.shape)
```

```python
# Encode layers
y_train = keras.utils.np_utils.to_categorical(y_train, num_classes=10)
y_test = keras.utils.np_utils.to_categorical(y_test, num_classes=10)
# print(y_train.shape)
# print(y_test.shape)
```

```python
# Extra libraries
import time
from keras.models import Sequential
from keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam, SGD
```

```python
models_dict = {}
```

```python
# Initialize the keras model
model = Sequential(name = "seq_1relu64_2softmax10_SGD")
model.add(Dense(64, activation = 'relu', input_dim = 784))
model.add(Dense(10, activation = 'softmax'))
model.summary() 

optimizer = SGD(learning_rate = 1e-3)
model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"])

# Fit the model
print('\n\nFit process:')
start = time.time()
model.fit(X_train, y_train, batch_size=32, epochs=20, verbose=2)
end = time.time()
print('Fit duration: %.6f ms' % (1000.0*(end-start)))

# Evaluate the model with train dataset
print('\n\nEvaluation train dataset process:')
evaluate_train = model.evaluate(X_train, y_train, verbose=1)

# Evaluate the model with test dataset
print('\n\nEvaluation test dataset process:')
evaluate_test = model.evaluate(X_test, y_test, verbose=1)

temp_list = [(end-start)*1000.0, evaluate_train[0], evaluate_train[1], evaluate_test[0], evaluate_test[1]]
# Save the results
models_dict[model.name] = temp_list
```

```python
neurons = [32, 64, 128] 
epochs = [10, 20, 30]

for neuron in neurons:
  for epoch in epochs:
    # Initialize the keras model
    model = Sequential(name = "seq_1relu" + str(neuron) + "_2softmax10_ADAM_ep" + str(epoch))
    model.add(Dense(neuron, activation = 'relu', input_dim = 784))
    model.add(Dense(10, activation = 'softmax'))
    model.summary() 

    optimizer = Adam(learning_rate = 1e-3)
    model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"])

    # Fit the model
    print('\n\nFit process:')
    start = time.time()
    model.fit(X_train, y_train, batch_size=32, epochs=epoch, verbose=2)
    end = time.time()
    print('Fit duration: %.6f ms' % (1000.0*(end-start)))

    # Evaluate the model with train dataset
    print('\n\nEvaluation train dataset process:')
    evaluate_train = model.evaluate(X_train, y_train, verbose=1)

    # Evaluate the model with test dataset
    print('\n\nEvaluation test dataset process:')
    evaluate_test = model.evaluate(X_test, y_test, verbose=1)

    temp_list = [(end-start)*1000.0, evaluate_train[0], evaluate_train[1], evaluate_test[0], evaluate_test[1]]
    # Save the results
    models_dict[model.name] = temp_list
```

```python
neurons = [32, 64, 128, 256, 512] 
loss_fun = ["MSE", "categorical_crossentropy"]
batch_size = [32, 64, 128]

for loss in loss_fun:
  for batch in batch_size:
    for neuron in neurons:
      # Initialize the keras model
      model = Sequential(name = "seq_1relu" + str(neuron) + "_2softmax10_ADAM_" + str(loss) + "_bs" + str(batch))
      model.add(Dense(neuron, activation = 'relu', input_dim = 784))
      model.add(Dense(10, activation = 'softmax'))
      model.summary() 

      optimizer = Adam(learning_rate = 1e-3)
      model.compile(optimizer=optimizer, loss=loss, metrics=["accuracy"])

      # Fit the model
      print('\n\nFit process:')
      start = time.time()
      model.fit(X_train, y_train, batch_size=batch, epochs=20, verbose=2)
      end = time.time()
      print('Fit duration: %.6f ms' % (1000.0*(end-start)))

      # Evaluate the model with train dataset
      print('\n\nEvaluation train dataset process:')
      evaluate_train = model.evaluate(X_train, y_train, verbose=1)

      # Evaluate the model with test dataset
      print('\n\nEvaluation test dataset process:')
      evaluate_test = model.evaluate(X_test, y_test, verbose=1)

      temp_list = [(end-start)*1000.0, evaluate_train[0], evaluate_train[1], evaluate_test[0], evaluate_test[1]]
      # Save the results
      models_dict[model.name] = temp_list
```

```python
# Print out the results
for (name, model_res) in models_dict.items():
  print("Model name:", name)
  print("-Train duration: %.2f ms" % model_res[0])
  print("-Train loss: %.4f %%" % (100*model_res[1]))
  print("-Train acc: %.4f %%" % (100*model_res[2]))
  print("-Test loss: %.4f %%" % (100*model_res[3]))
  print("-Test acc: %.4f %%\n" % (100*model_res[4]))
```

```python
best_test_acc_name = ''
best_test_acc_score = 0
# find best test accuracy
for (name, model_res) in models_dict.items():
   if model_res[4] > best_test_acc_score:
     best_test_acc_score = model_res[4]
     best_test_acc_name = name

print("Best model name (according to test acc): " + best_test_acc_name)
print("-Test acc: %.4f %%"  % (100*best_test_acc_score))
```

```python
# create some graphs
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

timeArray = []
trainAccArray = []
testAccArray = []

for (name, model_res) in models_dict.items():
  timeArray.append(model_res[0])
  trainAccArray.append(model_res[2])
  testAccArray.append(model_res[4])

window_size = 5
neurons = [32, 64, 128, 256, 512] 
loss_fun = ["MSE", "Categorical_crossentropy"]
batch_size = [32, 64, 128]

i = 0 
loss_fun_idx = 0
batch_idx = 0
while i < len(timeArray):

  fig, ax = plt.subplots(1, 3)

  if i > 14:
    loss_fun_idx = 1
  
  batch_idx = (int)(i / 5) % 3

  title = "Figure with " + loss_fun[loss_fun_idx] + " and batch_size=" + str(batch_size[batch_idx])
  fig.suptitle(title, fontsize=12, y=0.98)

  x1 = timeArray[i:i+window_size]
  x2 = trainAccArray[i:i+window_size]
  x3 = testAccArray[i:i+window_size]

  # Add 3 subplots within 1 Figure: 
  ax[0].set_ylabel('Time (ms)')
  ax[0].set_xlabel('Num of neurons')
  ax[0].set_title('Train Duration')
  ax[0].plot(neurons, x1, marker=".")

  ax[1].set_ylabel('Accuracy (%)')
  ax[1].set_xlabel('Num of neurons')
  ax[1].set_title('Train Accuracy')
  ax[1].plot(neurons, x2, marker=".") 

  ax[2].set_ylabel('Accuracy (%)')
  ax[2].set_xlabel('Num of neurons')
  ax[2].set_title('Test Accuracy')
  ax[2].plot(neurons, x3, marker=".") 

  fig.tight_layout(pad=1.0)

  fig.subplots_adjust(top=0.8)
  # fig.suptitle("\n".join([name]), y=0.98)

  plt.show()

  i += window_size
```
